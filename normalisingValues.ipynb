````python
def convertCSV(inputFile, outputFile):
    
    # Open file
    dataIn = open(inputFile, "r")
    dataOut = open(outputFile, "w")
    
    with dataIn as dataInRead:
        readValues = csv.reader(dataInRead, delimiter=',')
        
        count = 0
        
        for row in readValues:
            if count == 0:
                dataOut.write("\"{}\", \"{}\", \"{}\", \"{}\", \"{}\", \"{}\"\n".format(row[0], row[1], row[2], row[3], row[4], row[5]))
            else:
                img = cv2.imread("/Users/Work/Documents/Conversion/data/dataset/" + row[0], cv2.IMREAD_UNCHANGED)
 
                # Get image dimentions
                h, w, c = img.shape
                
                # Set new values
                xMinNorm = (float(row[1]) / w)
                yMinNorm = (float(row[2]) / h)
                xMaxNorm = (float(row[3]) / w)
                yMaxNorm = (float(row[4]) / h)

                dataOut.write("\"{}\", {}, {}, {}, {}, \"{}\"\n".format(row[0], xMinNorm, yMinNorm, xMaxNorm, yMaxNorm, row[5]))
                
            count += 1
            print(row[0])
        
    dataIn.close()
    dataOut.close()
````
